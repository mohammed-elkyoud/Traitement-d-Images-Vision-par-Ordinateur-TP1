```python
from PIL import Image
import numpy as np
path1= "dog.png"
path2= "dog1.png"
img1 = Image.open(path1).convert('L')
img2 = Image.open(path2).convert('L')
img1.show()
img2.show()
```

```python
img3= img1.resize((600,600))
img4 = img2.resize((600,600))
mat1 = np.array(img3)
mat2 = np.array(img4)
max_array = np.maximum(mat1, mat2)
max_image = Image.fromarray(max_array)
max_image.show()
```

## Exercice2

```python
threshold = 128
for i in range(max_array.shape[0]):
    for j in range(max_array.shape[1]):
        if max_array[i, j] > threshold:
            max_array[i, j] = 255
        else:
            max_array[i, j] = 0
img1_gray = Image.fromarray(max_array)
img1_gray.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image         
 # Définition de la matrice d'image (valeurs en niveaux de gris)
img = np.array([
    [25, 100, 220, 220, 200, 150],
    [200, 30, 30, 30, 220, 100],
    [220, 200, 30, 30, 100, 150],
    [150, 150, 150, 200, 150, 150],
    [100, 200, 30, 200, 150, 150],
    [100, 100, 100, 25, 25, 25]
], dtype=np.uint8)
  # Type uint8 pour une image en niveaux de gris
img             # Affichage de l'image avec Matplotlib
plt.imshow(img, cmap='gray')  # Utilisation d'une palette de gris
plt.colorbar()  # Ajouter une barre de couleur pour voir les valeurs des pixels
plt.title("Image sous forme de matrice")
plt.show()
```
